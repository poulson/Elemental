```python
from IPython.parallel import Client
c = Client(profile='mpi')
view = c[:]
view.activate()
view.block = True
%matplotlib inline
```

```python
%%px
# Initialize Elemental and print the process grid dimensions
import El
grid = El.DefaultGrid()
if El.mpi.WorldRank() == 0:
    print "Grid is", grid.Height(), "x", grid.Width()
```

```python
%%px
# Generate and display a Fox-Li/Landau matrix
n = 200
A = El.DistMatrix(El.zTag)
El.FoxLi(A,n)
El.Display(A,"Fox-Li/Landau matrix")
```

```python
%%px
# Display its spectral portrait
import math
realRes = imagRes = 100
portrait, box = El.SpectralPortrait(A,realRes,imagRes)
El.DisplayPortrait(portrait,box,"spectral portrait of Fox-Li/Landau matrix")
```

```python
%%px
# Display its singular values
s = El.SVD(A)
El.EntrywiseMap(s,math.log10)
El.Display(s,"log10(svd(A))")
```
